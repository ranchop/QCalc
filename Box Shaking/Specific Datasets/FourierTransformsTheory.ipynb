```python
import matplotlib.pyplot as plt
import therpy as tp
import numpy as np
%matplotlib inline
import pandas as pd

twopi = 2 * np.pi
```

# Spatial FT of Ideal Box Modes 

```python
L = 100
mode_num = 2

wl = 2*L / mode_num
k0 = twopi / wl

x = np.linspace(0, L, 200)
y = np.cos(k0 * x)

k = np.arange(0., 5*twopi/L, 0.02*twopi/L)
k, fk = tp.fourier_transform(x, y, k)
kscale = np.pi/L

fig, ax = plt.subplots(nrows=2, figsize=[5,7])
ax[0].plot(x, y, 'k')
ax[1].plot(k/kscale, np.abs(fk), 'k')
ax[1].plot(k/kscale, np.real(fk), '--')
ax[1].plot(k/kscale, np.imag(fk), '--')
ax[1].axvline(k0/kscale, c='k', linestyle='-', alpha=0.5)
ax[1].axhline(0, c='k', linestyle='-', alpha=0.5)
ax[1].set_xlabel(r'$k \: (\pi / L)$', fontsize=15)
plt.tight_layout()
```
